```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
X = np.random.rand(50,1)*100
y = 3.5 + X*np.random.rand(50,1)*20
print(y)
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
pred = model.predict(X)
```

```python
print(pred)
```

```python
Error = y - pred
```

```python
plt.plot(Error)
```

```python
plt.figure()
plt.scatter(X,y, color="blue", label="Data Points")
plt.plot(X,pred, color="red", linewidth=2, label="Regression Line")
plt.title("Linear Regression on random datasets")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("Slope (Coefficient):", model.coef_[0][0])
print("Intercept", model.intercept_[0])
```
