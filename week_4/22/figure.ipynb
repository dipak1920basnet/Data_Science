```python
import matplotlib.pyplot as plt
```

```python
# Create a figure with basic size and background color
fig = plt.figure(figsize=(10,8),facecolor="lightblue")

ax = fig.add_axes([0.1,0.1,0.8,0.8])
x = [1,2,3,4]
y = [1,2,3,4]
ax.plot(x,y)


plt.title("Simple line plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-axis")
plt.show()
```
