```python
# Import necessary library. 
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {"Age":[22, 25, 28, 24, 23, 29, 150, 27, 26, 22]}

df = pd.DataFrame(data)

plt.figure(figsize=(20,8))
#Box plot 
sns.boxplot(x=df["Age"])
plt.title("Box Plot for Age")
plt.xticks(np.arange(min(data["Age"]), max(data["Age"])+10,8))
plt.show()
```

Here the Box plot represents the interqualie range with a line for the median. 

2) Scatter Plot.

```python
sns.scatterplot(x=range(len(df["Age"])), y = df["Age"])
plt.title("Scatter Plot for Age")
plt.show()
```

The plot displays the data point along the y-axis. the point 150 is clearly a outliers as it is standing out from others.
