```python
import pandas as pd
```

```python
data1 = {'Name':['Jai', 'Anuj', 'Jai', 'Princi',
                 'Gaurav', 'Anuj', 'Princi', 'Abhi'],
        'Age':[27, 24, 22, 32,
               33, 36, 27, 32],
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj',
                   'Jaunpur', 'Kanpur', 'Allahabad', 'Aligarh'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd',
                         'B.Tech', 'B.com', 'Msc', 'MA']}
```

```python
import numpy as np
```

```python
# Applying the functions
```

```python
# for single grouping
datas = df.groupby(["Name"])

datas.aggregate([np.sum])
```

```python
# for grouping with multiples with keys

new_datas = df.groupby(["Name","Qualification"])

new_datas.aggregate([np.sum])
```

```python
# apply multiples function at once
datas["Age"].aggregate([np.sum, np.average, np.std])
```

```python
# for multiple keys and multiple functions
new_datas["Age"].aggregate([np.sum, np.average, np.std])
```

```python
# Now working with transformations
```

```python
df2 = pd.DataFrame(data1)
```

```python
grp2 = df2.groupby(["Name"])
```

```python
grp2.groups
```

```python
sc = lambda x : (x - x.mean())/x.std()

grp2['Age'].transform(sc)
```

```python
# Filterations

# Its a process used to discard the group based on a conditions

grp2 = df2.groupby(["Name"])

grp2.filter(lambda x: len(x) >= 2)
```

```python
# Results are combined back to dataframe after applying functions. 

df.groupby("Name").agg({"Age":"sum"})
```
