```python
# Import necessary modules
import numpy as np 
import pandas as pd 
from scipy import stats
```

```python
data =  {'Age': [22, 25, 28, 24, 23, 29, 150, 27, 26, 22]}
df = pd.DataFrame(data)
```

```python
z = np.abs(stats.zscore(df["Age"]))
print(z)
```

```python
print("Outliers: ")
outliers = df[z>3]
print(outliers)
```

#Interquartile-Range

```python
Q1 = df["Age"].quantile(0.25)
Q3 = df["Age"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1  - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outlier = df[(df["Age"] < lower_bound) | (df["Age"] > upper_bound)]
```

```python
print(outlier)
```

Handling Outliers

1) Removing outliers

```python
# we remove outliers where there is data entry error or measurement mistakes
df_cleaned = df[(df["Age"]>= lower_bound) & (df["Age"] <= upper_bound)]
```

```python
print("Cleaned Data: \n",df_cleaned)
```
