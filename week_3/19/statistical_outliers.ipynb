```python
# Import necessary modules
import numpy as np 
import pandas as pd 
from scipy import stats
```

```python
data =  {'Age': [22, 25, 28, 24, 23, 29, 150, 27, 26, 22]}
df = pd.DataFrame(data)
```

```python
z = np.abs(stats.zscore(df["Age"]))
print(z)
```

```python
print("Outliers: ")
outliers = df[z>3]
print(outliers)
```
