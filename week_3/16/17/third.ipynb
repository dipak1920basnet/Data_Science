```python
# Update Operations: MOdifying data in Pandas
```

```python
mini_data = {

    "Name":['Eve', 'Jack', 'Charlie', 'Henry', 'John'],
    "Age":[25, 30, 35, 40, 45],
    "City":['NY', 'LA', 'SF', 'Houston', 'Seattle'],
}

df = pd.DataFrame(mini_data)
```

```python
df.head()
```

```python
df.loc[df["Name"]=="Jack", "Age"] = 42
```

```python
df.head()
```

```python
df["City"] = ['Boston', 'Chicago', 'LA', 'Austin', 'Miami']
```

```python
df.head()
```

```python
# Update based on conditions:

df.loc[df["City"] == "LA","Age"] = 31
```

```python
df.head()
```
